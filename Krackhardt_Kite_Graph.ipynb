# Krackhardt Kite Graph

We'll generate a dataset to use for our Krackhardt Kite Graph and display it with two layouts.

 The Krackhardt Kite is a specific social network graph consisting of 10 nodes (actors) and 18 edges. Introduced by researcher David Krackhardt, this graph is designed to illustrate and distinguish between different network metrics, specifically degree, betweenness, and closeness centrality (*Social Analysis for Startups, p. 27*)



```python
#@title Generate Graph Dataset
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

# Generate the Krackhardt Kite graph
kite_graph = nx.krackhardt_kite_graph()

# Map names to the nodes
mapping = {
    0: 'Andre', 1: 'Beverley', 2: 'Carol', 3: 'Diane',
    4: 'Ed', 5: 'Fernando', 6: 'Garth', 7: 'Heather',
    8: 'Ike', 9: 'Jane'}
kite_graph = nx.relabel_nodes(kite_graph, mapping)

# Create a dataset from the edges
edges = list(kite_graph.edges())
df_kite = pd.DataFrame(edges, columns=['Source', 'Target'])

print("Generated Dataset (First 5 rows):")
display(df_kite)
```

```python
#@title Generate Krackhardt Kite Graph

# Visualize the Krackhardt Kite Graph
plt.figure(figsize=(10, 6))

# Define the spring_layout and kamada_kawai_layout for graph
pos1 = nx.kamada_kawai_layout(kite_graph)
pos2 = nx.spring_layout(kite_graph)

# Display Kamada Kawai Layout
nx.draw(kite_graph, pos1, with_labels=True, node_color='skyblue',
        edge_color='gray', node_size=2500, font_size=10, font_weight='bold')
plt.title("The Krackhardt Kite Graph (Kamada Layout)")
plt.show()

# Display Spring Layout
nx.draw(kite_graph, pos2, with_labels=True, node_color='skyblue',
        edge_color='gray', node_size=2500, font_size=10, font_weight='bold')
plt.title("The Krackhardt Kite Graph (Spring Layout)")
plt.show()
```
